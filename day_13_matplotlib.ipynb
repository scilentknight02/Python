# Matplot

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,100)
print(x)
print()
y = np.random.randint(1, 50, 100)
print(y)
plt.plot(x,y, marker='p') # marker range 1 - 11 and 'o' 'p' etc
plt.xlabel("height")
plt.ylabel("width")
plt.grid()
```

```python
x = np.linspace(10,20)
y = np.linspace(20,30)
print(x)
print(y)
a = np.linspace(10,20)
b = np.linspace(10,20)
print(a)
print(b)
plt.plot(x,y)
plt.plot(a,b)
plt.xlabel("Day")
plt.ylabel("Hours")
plt.legend("Study Time \n Screen Time")
```

```python
activities = ['sleep', 'study', 'play', 'eat', 'gaming']
hours = [8, 5, 2, 2, 4]
colors = ['#aaa', '#faf', '#dda', '#bbb', '#ccc']
plt.pie(hours, labels=activities, colors=colors)
plt.title('Daily Activity Breakdown')
plt.axis('equal')
plt.savefig('plot1.png')
plt.show()
```

```python
activities = ['sleep', 'study', 'play', 'eat', 'gaming']
hours = [8, 5, 2, 2, 4]
colors = ['#aaa', '#faf', '#dda', '#bbb', '#ccc']
plt.bar(activities, hours, color='orange')
plt.title('Daily Activity Breakdown')
plt.axis('equal')
plt.savefig('plot1.png')
plt.show()
```
